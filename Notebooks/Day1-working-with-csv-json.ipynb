Working on JSON
JSON is a lightweight data format that stores data in key-value pairs. It's commonly used in APIs and can handle nested structures.

```python
# 1. Basic read (default)
pd.read_json('../Datasets/train.json')
```

```python
# 2. Explicitly specify orient
pd.read_json('../Datasets/train.json', orient='records')
```

```python
# 3. Check the JSON file structure
with open('../Datasets/train.json', 'r') as f:
    first_chars = f.read(200)
    print(first_chars)
```

```python
# 4. Specify data types for columns (saves memory)
pd.read_json(
    '../Datasets/train.json',
    dtype={'id': 'int32', 'cuisine': 'category'}
)
```

```python
# 5. Specify encoding
pd.read_json(
    '../Datasets/train.json',
    encoding='utf-8'
)
```

```python
# 6. Handle encoding errors
pd.read_json(
    '../Datasets/train.json',
    encoding='utf-8',
    encoding_errors='ignore'  # or 'replace', 'strict'
)
```

```python
# 7. High precision floats (if you have decimal numbers)
pd.read_json(
    '../Datasets/train.json',
    precise_float=True
)
```

```python
# 8. Specify return type
pd.read_json(
    '../Datasets/train.json',
    typ='frame'  # 'frame' for DataFrame (default), 'series' for Series
)
```
