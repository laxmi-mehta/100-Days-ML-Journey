```python
import pandas as pd
```

```python
df = pd.read_csv('../Datasets/train (1).csv')
df.head()
```

1. How big is the data?

```python
# Get dataset dimensions
rows, columns = df.shape

print("=" * 50)
print("üìè DATASET DIMENSIONS")
print("=" * 50)
print(f"Total Rows (Records): {rows}")
print(f"Total Columns (Features): {columns}")
print(f"Total Data Points: {rows * columns:,}")
print("=" * 50)
```

8. Column Names

```python
# Display all column names
print("üìã Column Names:\n")
print(df.columns.tolist())
print(f"\n‚úÖ Total {len(df.columns)} columns")
```

9. Data Types Summary

```python
# Count of each data type
print("üìä Data Type Distribution:\n")
print(df.dtypes.value_counts())
```

10. Detailed Statistics

```python
# More detailed statistics including percentiles
print("üìä Detailed Statistical Summary:\n")
df.describe(percentiles=[.1, .25, .5, .75, .9, .95, .99])
```

11. Statistics for All Columns

```python
# Include categorical columns in summary
print("üìã Statistical Summary (All Features):\n")
df.describe(include='all')
```

12. Unique Value Counts

```python
# Count unique values in each column
print("üî¢ Unique Value Counts:\n")
unique_counts = df.nunique().sort_values(ascending=False)
print(unique_counts)
```

13. Mean, Median, Mode

```python
# Calculate mean, median, mode for numerical columns
print("üìà Central Tendency Measures:\n")
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in numerical_cols:
    print(f"\n{col}:")
    print(f"  Mean: {df[col].mean():.2f}")
    print(f"  Median: {df[col].median():.2f}")
    print(f"  Mode: {df[col].mode().values[0] if len(df[col].mode()) > 0 else 'N/A'}")
```

14. Standard Deviation & Variance

```python
# Calculate standard deviation and variance
print("üìä Dispersion Measures:\n")
for col in numerical_cols:
    print(f"\n{col}:")
    print(f"  Std Deviation: {df[col].std():.2f}")
    print(f"  Variance: {df[col].var():.2f}")
    print(f"  Range: {df[col].max() - df[col].min():.2f}")
```

15. Skewness & Kurtosi

```python
# Calculate skewness and kurtosis
print("üìâ Distribution Shape:\n")
for col in numerical_cols:
    print(f"\n{col}:")
    print(f"  Skewness: {df[col].skew():.2f}")
    print(f"  Kurtosis: {df[col].kurtosis():.2f}")
```
