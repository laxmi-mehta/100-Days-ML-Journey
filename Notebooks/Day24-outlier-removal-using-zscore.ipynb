Z-score measures how far a data point is from the mean in terms of standard deviations; points with |Z| > threshold (commonly 3) are often treated as outliers. Best when the feature distribution is roughly normal (bell-shaped)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print("Mean value of cgpa",df['cgpa'].mean())
print("Std value of cgpa",df['cgpa'].std())
print("Min value of cgpa",df['cgpa'].min())
print("Max value of cgpa",df['cgpa'].max())
```

```python
# Finding the boundary values
print("Highest allowed",df['cgpa'].mean() + 3*df['cgpa'].std())
print("Lowest allowed",df['cgpa'].mean() - 3*df['cgpa'].std())
```

```python
# Finding the outliers
df[(df['cgpa'] > 8.80) | (df['cgpa'] < 5.11)]
```

Trimming: Completely removes outlier rows from the dataset based on chosen criteria (e.g., beyond Z-score or percentile thresholds). Good when you believe outliers are errors or irrelevant.

```python
# Trimming

new_df = df[(df['cgpa'] < 8.80) & (df['cgpa'] > 5.11)]
new_df
```

```python
# Approach 2

# Calculating the Zscore

df['cgpa_zscore'] = (df['cgpa'] - df['cgpa'].mean())/df['cgpa'].std()
```

```python
df.head()
```

```python
df[df['cgpa_zscore'] > 3]
```

```python
df[df['cgpa_zscore'] < -3]
```

```python
df[(df['cgpa_zscore'] > 3) | (df['cgpa_zscore'] < -3)]
```

```python
# Trimming 
new_df = df[(df['cgpa_zscore'] < 3) & (df['cgpa_zscore'] > -3)]
```

```python
new_df
```

Capping (Winsorization): Replaces outliers with maximum/minimum threshold values (e.g., top 5% capped to 95th percentile). Keeps all rows but limits their impact on learning.

```python
upper_limit = df['cgpa'].mean() + 3*df['cgpa'].std()
lower_limit = df['cgpa'].mean() - 3*df['cgpa'].std()
```

```python
lower_limit
```

```python
df['cgpa'] = np.where(
    df['cgpa']>upper_limit,
    upper_limit,
    np.where(
        df['cgpa']<lower_limit,
        lower_limit,
        df['cgpa']
    )
)
```
